```python
from matplotlib import pyplot as plt
import csv
import pandas as pd
import io
from google.colab import files
cars=files.upload()
```

```python
data=pd.read_csv(io.BytesIO(cars['mtcars.csv']))
print(data)
```

```python
hp = data['hp']
plt.hist(hp,edgecolor='black')
plt.title('Horse Power of Cars')
plt.xlabel('Horse Power')
plt.ylabel('Number of cars')
plt.tight_layout()
```

```python
x_label=data['mpg']
y_label=data['model']
plt.bar(x_label,y_label)
plt.show()
```
